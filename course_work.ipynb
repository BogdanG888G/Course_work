```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

countries = ['Таиланд', 'Малайзия', 'Индонезия', 'Индия', 'Мексика', 'Турция', 'Румыния',
'Перу', 'Бразилия', 'Сингапур', 'Колумбия', 'Южная Африка', 'Южная Корея', 'Италия', 'Среднее по странам',
'Аргентина', 'Чили', 'Польша', 'Испания', 'Венгрия', 'Германия', 'Великобритания', 'Нидерланды',
'Япония', 'Новая Зеландия', 'Австралия', 'Швеция', 'Бельгия', 'Канада', 'Ирландия', 'Франция', 'Соединенные Штаты']
values = [0.73, 0.7, 0.69, 0.67, 0.66, 0.65, 0.62, 0.6, 0.6, 0.57, 0.56, 0.55,
          0.55, 0.53, 0.52, 0.51, 0.5, 0.5, 0.49, 0.46, 0.45, 0.45, 0.44, 0.44, 0.43, 0.42, 0.42,
          0.39, 0.39, 0.39, 0.37, 0.36]

# Выбираем только нужные данные (31 страну)
df = pd.DataFrame({'Name': countries, 'Value': values})

# Сортируем данные по убыванию
df = df.sort_values(by='Value', ascending=False)

# initialize the figure
plt.figure(figsize=(24, 14))
ax = plt.subplot(111, polar=True)
plt.axis('off')

# Constants = parameters controlling the plot layout:
upperLimit = 100
lowerLimit = 30
labelPadding = 8

# Compute max and min in the dataset
max_value = df['Value'].max()

# Let's compute heights: they are a conversion of each item value in those new coordinates
slope = (upperLimit - lowerLimit) / max_value
heights = slope * df.Value + lowerLimit

# Compute the width of each bar. In total we have 2*Pi = 360°
width = 2 * np.pi / len(df.index)

# Compute the angle each bar is centered on:
indexes = list(range(len(df.index)))
angles = [element * width for element in indexes]

# Draw bars
bars = ax.bar(
    x=angles,
    height=heights,
    width=width,
    bottom=lowerLimit,
    linewidth=2,
    edgecolor="white",
    color=['orange' if country == 'Среднее по странам' else color1 for country in df['Name']]
)

# Add labels
for bar, angle, height, label, value in zip(bars, angles, heights, df["Name"], df['Value']):
    # Labels are rotated. Rotation must be specified in degrees :(
    rotation = np.rad2deg(angle)
    # Flip some labels upside down
    alignment = ""
    if angle >= np.pi / 2 and angle < 3 * np.pi / 2:
        alignment = "right"
        rotation = rotation + 180
    else:
        alignment = "left"
    # Finally add the labels
    ax.text(
        x=angle,
        y=lowerLimit + bar.get_height() + labelPadding,
        s=f'{label} ({value * 100:.0f})',
        ha=alignment,
        va='center',
        rotation=rotation,
        rotation_mode="anchor",
        color='white',
        fontsize=16  # Увеличенный размер шрифта
    )
plt.savefig('round.png', transparent = True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Исходные данные
countries = ['Таиланд', 'Индонезия', 'Мексика', 'Индия', 'Малайзия', 'Румыния', 'Перу',
             'Венгрия', 'Италия', 'Южная Африка', 'Турция', 'Бразилия', 'Польша', 'Сингапур', 'Колумбия',
             'Глобальный', 'Испания', 'Нидерланды', 'Аргентина', 'Чили', 'Германия', 'Новая Зеландия', 'Великобритания',
             'Бельгия', 'Южная Корея', 'Швеция', 'Ирландия', 'Австралия', 'Канада', 'Япония', 'США',
             'Франция']
values = [0.72, 0.68, 0.66, 0.64, 0.61, 0.61, 0.6, 0.6, 0.59, 0.58, 0.57, 0.56, 0.55,
          0.54, 0.53, 0.5, 0.49, 0.47, 0.45, 0.44, 0.44, 0.42, 0.42, 0.41, 0.4, 0.39, 0.38, 
          0.38, 0.34, 0.32, 0.32, 0.32]

# Создаем DataFrame
df = pd.DataFrame({'Country': countries, 'Value': values})

# Вычисляем отклонение от 0.5 (50%)
df['Deviation'] = df['Value'] - 0.5

# Определяем цвета: зеленый для значений выше 0.5, красный для значений ниже 0.5
df['Color'] = np.where(df['Deviation'] > 0, 'green', 'red')

# Сортируем значения по отклонению
df.sort_values('Deviation', inplace=True)
df.reset_index(inplace=True, drop=True)

# Строим график
plt.figure(figsize=(16, 12), dpi=80)
plt.hlines(y=df.index, xmin=0, xmax=df['Deviation'], color=df['Color'], alpha=0.6, linewidth=5)

# Декорации
plt.yticks(df.index, df['Country'], fontsize=14, color='white')
plt.axvline(x=0, color='white', linestyle='--', alpha=0.5)  # Вертикальная линия на 50%
plt.grid(linestyle='--', alpha=0.5)
plt.gca().axes.xaxis.set_visible(False)  # Убираем ось x

# Подписи значений
for i, (deviation, value) in enumerate(zip(df['Deviation'], df['Value']*100)):
    plt.text(deviation, i, f'{value:.0f}%', 
             horizontalalignment='right' if deviation < 0 else 'left', 
             verticalalignment='center', 
             fontdict={'color': 'white', 'size': 12})

plt.savefig('dif.png', transparent = True)
plt.box(False)
plt.grid(False)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Создаем DataFrame
df = pd.DataFrame({'Лучше': [51, 39, 37, 34, 32],
                   'Останется также': [29, 38, 39, 30, 23],
                   'Не знаю': [9, 11, 8, 13, 10],
                   'Хуже': [11, 12, 16, 23, 35]},
                  index=[1231, 2213, 3213, 4321, 5123])

fig, ax = plt.subplots(figsize=(10, 6))

# Задаем цвет для color1
color1 = '#7F00FF'

# Построение горизонтального stacked bar chart
bars = df.plot(kind='barh', stacked=True, color=[color1, '#1EFAF2', '#E5FA1E', '#FA6B1E'], ax=ax)

# Удаляем сетку
ax.grid(False)

# Удаляем рамку (spines)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Убираем ось x
plt.gca().axes.xaxis.set_visible(False)
plt.gca().axes.yaxis.set_visible(False)

# Добавляем значения в белом цвете
for container in bars.containers:
    # Добавляем значения к каждому столбцу
    ax.bar_label(container, label_type='center', color='black')

# Расположение легенды под графиком
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=2, labelcolor = 'white', facecolor = color1)

ax.yaxis.label.set_color('violet')
# Корректировка макета для учета легенды
plt.subplots_adjust(bottom=0.2)
# Показываем график
plt.savefig('refnv.png', transparent=True)
plt.show()
```
